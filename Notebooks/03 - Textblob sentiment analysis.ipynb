```python
import pandas as pd

from textblob import TextBlob
```

```python
df = pd.read_csv('first_week_oct_2015_comments_by_top_400_with_scores.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
tb_sentiment = df['text'].apply(lambda text: TextBlob(text).sentiment)
```

```python
tb_sentiment[0]
```

```python
df[['tb_polarity', 'tb_subjectivity']] = tb_sentiment.apply(pd.Series)
```

```python
df.head()
```

```python
for text in df.sort_values(by='tb_polarity')['text'].head(20):
    print(text, '\n')
```

```python
df.to_csv('first_week_oct_2015_comments_by_top_400_with_scores_v2.csv', index=False)
```
